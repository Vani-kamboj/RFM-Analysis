```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('C:/Users/hp/Downloads/All Payments (1).csv')
```

```python
df
```

## Recency

```python
df['Date']=pd.to_datetime(df['created_at']).dt.date

df.Date
```

```python
recency = df.groupby(by='email',
                        as_index=False)['Date'].max()
recency.columns = ['email', 'Date']
recent_date = recency['Date'].max()
recency['Recency'] = recency['Date'].apply(
    lambda x: (recent_date - x).days)
recency.head()
```

## Frequency

```python
frequency = df.drop_duplicates().groupby(
    by=['email'], as_index=False)['Date'].count()
frequency.columns = ['email', 'Frequency']
frequency.head()
```

## Monetry Value

```python
monetary = df.groupby(by='email', as_index=False)['amount'].sum()
monetary.columns = ['email', 'Monetary']
monetary.head()
```

```python
rf=recency.merge(frequency , on='email')
rfm=rf.merge(monetary, on='email')
rfm
```

```python
rfm['Recency_rank'] = rfm['Recency'].rank(ascending=False)
rfm['Frequency_rank'] = rfm['Frequency'].rank(ascending=True)
rfm['Monetary_rank'] = rfm['Monetary'].rank(ascending=True)
```

```python
rfm['R_rank'] = (rfm['Recency_rank']/rfm['Recency_rank'].max())*100
rfm['F_rank'] = (rfm['Frequency_rank']/rfm['Frequency_rank'].max())*100
rfm['M_rank'] = (rfm['Monetary_rank']/rfm['Monetary_rank'].max())*100
 
rfm.drop(columns=['Recency_rank', 'Frequency_rank', 'Monetary_rank'], inplace=True)
 
rfm.head()
```

```python
rfm['rfm_Score'] = 0.15*rfm['R_rank']+0.28 * \
    rfm['F_rank']+0.57*rfm['M_rank']
rfm['rfm_Score'] *= 0.05
rfm = rfm.round(2)
rfm[['email', 'rfm_Score']].head(10)
```
